```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
colores = {e: 'red' if v == 1 else 'green' for e, v in nx.get_node_attributes(EliteNet, "multi").items()}
```

```python
nx.set_node_attributes(EliteNet, colores, "color")
```

```python
node_colors = [EliteNet.nodes[node]['color'] for node in EliteNet.nodes]
```

```python
nx.draw_random(EliteNet, node_color=node_colors, edge_color='lightblue', with_labels=True, font_size=8)
plt.show()
```

```python
import networkx as nx
```

```python
from pyvis.network import Network
```

```python
labels={i:str(i) for i in EliteNet.nodes()}
labels # string as label
```

```python
nx.set_node_attributes(EliteNet,labels,'label') # values according to id
nx.set_node_attributes(EliteNet,'red','color') # for all
EliteNet.nodes.data()
```

```python
Gvis = Network('500px', '500px',notebook=True,cdn_resources='remote')
Gvis.from_nx(EliteNet.copy()) # create a copy!
Gvis.show('Gvis.html')
```
